```python
import numpy as np
import matplotlib.pyplot as plt

n = int(input())

x = np.linspace(1, n, n, dtype=(int))

y = x**2

plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='$y = x * x$')
plt.title('Plot of $y = x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
