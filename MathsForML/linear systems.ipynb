```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([
    [-1, 3],
    [3, 2]
], dtype=np.dtype(float))
b = np.array([
    [7],
    [1]
], dtype=np.dtype(float))

ans = np.linalg.solve(a, b)
print(ans)
```

```python
d = np.linalg.det(a)
print(d)
```

```python
a_system = np.hstack((
    a,
    b
))
print(a_system)
```

```python
plt.plot(a)
```

# No Solutions

```python
A_2 = np.array([
    [-1, 3],
    [3, -9]
], dtype=np.dtype(float))

B_2 = np.array([
    [7],
    [1]
], dtype=np.dtype(float))

print(A_2)
print(B_2)

d_2 = np.linalg.det(A_2)

if d_2 == 0:
    print("No Solutions!")
else:
    ans_2 = np.linalg.solve(A_2, B_2)
    print(ans_2)
```

```python
A_2_system = np.hstack((
    A_2,
    B_2.reshape(2, 1)
))
print(A_2_system)
```

```python
plt.plot(A_2_system)
```
